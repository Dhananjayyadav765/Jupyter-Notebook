```python
import pandas as pd
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/quote/TSLA/history?period1=1578700800&period2=1704931200&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true'
```

```python
data  = requests.get(url).text
print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
columns=["Date","Open","High","Low","Close","Volume"]
```

```python
tesla_revenue_data = pd.DataFrame(columns=["Date","Open","High","Low","Close","Volume"])
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    volume = col[5].text
tesla_revenue_data = tesla_revenue_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Volume":volume}, ignore_index=True)     
```

```python
tesla_revenue_data.tail()
```
